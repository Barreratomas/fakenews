```python
# 1. Restaurar Workspace
import os
import shutil
import sys

# Ruta donde Kaggle monta el dataset
INPUT_DIR = '/kaggle/input/fake-news-workspace'
WORKING_DIR = '/kaggle/working'

print('📂 Contenido del input:', os.listdir(INPUT_DIR))

# Copiar todo al directorio de trabajo (para tener permisos de escritura)
print('� Restaurando estructura del proyecto en /kaggle/working...')

# Kaggle a veces descomprime el zip, a veces no. Manejamos ambos casos.
zip_path = os.path.join(INPUT_DIR, 'fake_news_workspace.zip')
if os.path.exists(zip_path):
    print('   -> Descomprimiendo ZIP maestro...')
    shutil.unpack_archive(zip_path, WORKING_DIR)
else:
    print('   -> Copiando archivos planos...')
    # Si Kaggle ya lo descomprimió, copiamos recursivamente
    shutil.copytree(INPUT_DIR, WORKING_DIR, dirs_exist_ok=True)

print('✅ Estructura restaurada:')
for root, dirs, files in os.walk(WORKING_DIR):
    level = root.replace(WORKING_DIR, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    # Mostrar solo algunos archivos para no saturar el log
    for f in files[:5]:
        print(f'{subindent}{f}')
    if len(files) > 5:
        print(f'{subindent}... ({len(files)-5} más)')

# 2. Instalar Dependencias
if os.path.exists('requirements.txt'):
    print('⬇️ Instalando dependencias...')
    !pip install -r requirements.txt -q
else:
    print('❌ ALERTA: No se encontró requirements.txt')

# 3. Verificar Datos Procesados
if os.path.exists('data/processed/train'):
    print('✅ Datos procesados detectados. Saltando build_hf_datasets.py')
else:
    print('⚠️ No hay datos procesados. Ejecutando preprocesamiento...')
    !PYTHONPATH=. python src/preprocessing/build_hf_datasets.py

print('✅ Entorno listo para entrenar.')
```
